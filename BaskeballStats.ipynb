```python
from matplotlib.patches import Circle, Rectangle, Arc

def draw_court(ax=None, lw=2, boundaries=False):

    if ax is None:
        ax = plt.gca()
        
    Out = Rectangle((-250, -47.5), 500, 470, linewidth=lw, color='burlywood', fill = 'brown') 

    H = Circle((0, 0), radius=7.5, linewidth=lw, color='black', fill= None)

    BB = Rectangle((-30, -7.5), 60, -1, linewidth=lw, color='black')

    OB = Rectangle((-80, -47.5), 160, 190, linewidth=lw, color='black', fill=False)
    
    IB = Rectangle((-60, -47.5), 120, 190, linewidth=lw, color='black', fill=False)

    FTT = Arc((0, 142.5), 120, 120, theta1=0, theta2=180, linewidth=lw, color='black', fill=False)
    
    FTB = Arc((0, 142.5), 120, 120, theta1=180, theta2=0, linewidth=lw, color='black', linestyle='dashed')
    
    R = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw, color = 'black')

    C31 = Rectangle((-220, -47.5), 0, 140, linewidth=lw, color='black')
    C32 = Rectangle((220, -47.5), 0, 140, linewidth=lw, color='black')

    A = Arc((0, 0), 475, 475, theta1=22, theta2=158, linewidth=lw, color='black')


    court = [H, BB, OB, IB, FTT, FTB, R, C31, C32, A]

    if boundaries:
        boundaries = Rectangle((-250, -47.5), 500, 470, linewidth=lw, color='black', fill= False, label = 'James Harden')
        court.append(boundaries)

    for i in court:
        ax.add_patch(i)

    return ax
```
